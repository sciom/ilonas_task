# Data cleanup

The data is in the file `sound and tree data.xlsx`. 

First, we need to clean the data a bit. We will separate longitudes and latitudes from the `location` column and create new columns for them. we will also add a column for the distance between the tree and the speaker, aswell as a column for labels "tree", "amplifier" and "speaker". Reformating of the data will be done using python. The data will be saved in a new file `sound_and_tree_data_cleaned.csv` and provided with this document. Note beforehand we aded a new column namde label to the data, which is the label of the data, either "tree", "amplifier" or "speaker".


```python
import pandas as pd

data = pd.read_csv("sound_and_tree_data.csv")
data
```

```python

# Separating longitude and lattitude from the location column

data["longitude"] = data["location"].apply(lambda x: x.split(",")[0])
data["latitude"] = data["location"].apply(lambda x: x.split(",")[1])
    
        
```
